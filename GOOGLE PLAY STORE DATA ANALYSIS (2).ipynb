```python
import numpy as np

def convert_size(size):
    if isinstance(size, str):  # Ensure size is a string to avoid errors
        if 'M' in size:
            return float(size.replace('M', ''))
        elif 'k' in size:
            return float(size.replace('k', '')) / 1024
    return np.nan  # Return NaN for invalid values

apps['Size'] = apps['Size'].apply(convert_size)
```

```python
# Save each Plotly figure to an HTML file
def save_plot_as_html(fig, filename, insight):
    global plot_containers
    filepath = os.path.join(html_files_path, filename)
    html_content = pio.to_html(fig, full_html=False, include_plotlyjs='inline')
    # Append the plot and its insight to plot_containers
    plot_containers += f"""
    <div class="plot-container" id="{filename}" onclick="openPlot('{filename}')">
        <div class="plot">{html_content}</div>
        <div class="insights">{insight}</div>
    </div>
    """
    fig.write_html(filepath, full_html=False, include_plotlyjs='inline')
```

```python
# TASK 1
df = mergeddata


df = df[df['Reviews'].astype(float) > 1000]

# Step 2: Get the top 5 categories by app count
top_categories = df['Category'].value_counts().nlargest(5).index
df = df[df['Category'].isin(top_categories)]

# Step 3: Define rating groups
def rating_group(rating):
    try:
        rating = float(rating)
        if rating <= 3:
            return '1-3 stars'
        elif rating == 4:
            return '4 stars'
        elif rating == 5:
            return '5 stars'
        else:
            return 'Other'
    except:
        return 'Unknown'


df['Rating_Group'] = df['Rating'].apply(rating_group)

# Step 4: Group and count by Category, Rating_Group, and Sentiment
sentiment_counts = df.groupby(['Category', 'Rating_Group', 'Sentiment']).size().reset_index(name='Count')

# Combine Category and Rating Group for x-axis
sentiment_counts['Category & Rating Group'] = sentiment_counts['Category'] + ' - ' + sentiment_counts['Rating_Group']

# Create bar chart
fig11 = px.bar(
    sentiment_counts,
    x='Category & Rating Group',
    y='Count',
    color='Sentiment',
    title='Sentiment Distribution by Rating Group (Top 5 Categories, >1000 Reviews)',
    color_discrete_sequence=px.colors.sequential.Greens,
    width=400,
    height=300
)

fig11.update_layout(
    plot_bgcolor='black',
    paper_bgcolor='black',
    font_color='white',
    title_font={'size':16},
    xaxis=dict(title_font={'size':12}),
    yaxis=dict(title_font={'size':12}),
    margin=dict(l=10, r=10, t=30, b=10)
)

# Save to HTML
fig11.write_html("Rating Graph fig.11.html")
save_plot_as_html(fig11,"Rating Graph 11.html","stacked bar chart, segmented by rating groups (e.g., 1-2 stars, 3-4 stars, 4-5 stars). Include only apps with more than 1,000 reviews and group by the top 5 categories.")


```

```python
# TASK 2

from datetime import datetime
import pytz

# Load your Google Play Store data
df = mergeddata

# Filter data
filtered_df = df[
    (df['Rating'] >= 4.0) &
    (df['Size'] >= 10) &
    (df['Last Updated'].dt.month == 1)
]

# Get top 10 categories by installs
top_categories = (
    filtered_df.groupby('Category')['Installs']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .index
)

top_df = filtered_df[filtered_df['Category'].isin(top_categories)]

# Group for chart
summary = top_df.groupby('Category').agg({
    'Rating': 'mean',
    'Reviews': 'sum'
}).reset_index()

# Check current time in IST
ist = pytz.timezone('Asia/Kolkata')
now = datetime.now(ist)

if 15 <= now.hour < 17:
    
    fig12 = px.bar(
        summary,
        x='Category',
        y='Rating',
        color='Reviews',
        title='Top 10 App Categories by Installs (Filtered)',
        color_continuous_scale=px.colors.sequential.Greens,
        width=400,
        height=300
    )

    fig12.update_layout(
        plot_bgcolor='black',
        paper_bgcolor='black',
        font_color='white',
        title_font={'size': 16},
        xaxis=dict(title_font={'size': 12}),
        yaxis=dict(title_font={'size': 12}),
        margin=dict(l=10, r=10, t=30, b=10)
    )

    # Save and show plot
    fig12.write_html("Total install by cateogry .html")
    save_plot_as_html(fig12,"Total install by category","Filter out any categories where the average rating is below 4.0 and size below 10 M and last update should be Jan month")

    


else:
    print("Chart is not visible outside 3PM–5PM IST.")

```

```python
# TASK 3

from datetime import datetime
import pytz

# Use mergeddata as the source
df = mergeddata

# Standardize columns
df['App'] = df['App'].astype(str)
df['Category'] = df['Category'].astype(str)
df['Installs'] = df['Installs'].replace('[+,]', '', regex=True).astype(int)
df['Reviews'] = df['Reviews'].astype(int)
df['Last Updated'] = pd.to_datetime(df['Last Updated'])

# Apply Filters
df = df[
    (~df['App'].str.startswith(('X', 'Y', 'Z'))) &
    (~df['App'].str.contains('S', case=False)) &
    (df['Category'].str.startswith(('E', 'C', 'B'))) &
    (df['Reviews'] > 500)
]

# Group by month
df['Month'] = df['Last Updated'].dt.to_period('M').dt.to_timestamp()

# Aggregate installs
monthly = df.groupby(['Month', 'Category'])['Installs'].sum().reset_index()

# Translate categories
translation = {
    'Beauty': 'सौंदर्य',         # Hindi
    'Business': 'வணிகம்',       # Tamil
    'Dating': 'Partnersuche'    # German
}
monthly['Category'] = monthly['Category'].replace(translation)

# Sort & compute growth
monthly.sort_values(by=['Category', 'Month'], inplace=True)
monthly['Prev_Installs'] = monthly.groupby('Category')['Installs'].shift(1)
monthly['Growth'] = (monthly['Installs'] - monthly['Prev_Installs']) / monthly['Prev_Installs']

# IST time check
ist = pytz.timezone('Asia/Kolkata')
now = datetime.now(ist)

if 18 <= now.hour < 21:
    # Plot line chart
    fig13 = px.line(
        monthly,
        x='Month',
        y='Installs',
        color='Category',
        title='Total Installs Over Time by App Category',
        width=800,
        height=500
    )

    
    for cat in monthly['Category'].unique():
        subset = monthly[(monthly['Category'] == cat) & (monthly['Growth'] > 0.2)]
        fig13.add_scatter(
            x=subset['Month'],
            y=subset['Installs'],
            mode='lines',
            line=dict(width=0),
            fill='tozeroy',
            fillcolor='rgba(255,165,0,0.3)',
            name=f"{cat} >20% Growth"
        )

    
    fig13.update_layout(
        plot_bgcolor='black',
        paper_bgcolor='black',
        font_color='white',
        title_font={'size': 20},
        xaxis=dict(title='Month'),
        yaxis=dict(title='Total Installs'),
        margin=dict(l=20, r=20, t=50, b=20)
    )

    
    fig13.write_html("Total_Installs_By_Category.fig13.html")
    
    save_plot_as_html(fig13,"Total_Installs_By_Category.fig13.html","chart to show the trend of total installs over time, segmented by app category")

else:
    print("Chart is only visible between 6 PM and 9 PM IST.")
```
